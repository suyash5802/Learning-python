```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer

```

```python
df=pd.read_csv("HousePricePrediction.csv")
df
df.drop(["Id"],axis=1)
```

```python
df.isnull().sum()
spi=SimpleImputer(strategy="mean")
df[["SalePrice"]]=spi.fit_transform(df[["SalePrice"]])



```

```python
df.isnull().sum()
df.dropna()
```

```python
fea=["MSZoning","LotConfig","BldgType","Exterior1st"]
df=pd.get_dummies(df,columns=fea,drop_first=True)
```

```python
X=df.drop(["SalePrice"],axis=1)
y=df["SalePrice"]
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
X_train=spi.fit_transform(X_train)
X_test=spi.fit_transform(X_test)
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score,accuracy_score,recall_score
y_pred=model.predict(X_test)
print(r2_score(y_test,y_pred))
```
